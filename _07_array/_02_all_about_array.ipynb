1)  How to create and trasverse an Array

```python
from array import array

my_array = array("i", [1, 2, 3, 4, 5])

# Traversing now
for i in my_array:
    print(i, end=", ")
```

2. Ascessing Individual elements in array.

```python
print(f"Last element : {my_array[len(my_array) -1]}")
print(my_array)

#adding an element in last(append)
my_array.append(7)
print(my_array)

#add anywhere(insert)
my_array.insert(3, 9)
print(my_array)

#adding more than one elements(in last index)
my_array_01 =  array("i",[11, 12, 13])
my_array.extend(my_array_01)

#adding from the list type to array
my_list : list = [21, 22, 23]
my_array.fromlist(my_list)
print(my_array)
```

3. Removing or deleting elements from the array

```python
#using remove to remove the first found occurrence of the given element

my_array.remove(7)
print(my_array)


# Remove only the last element
my_array.pop()
print(my_array)
```

4) get the Index of element)

```python
# Return the index of first occurrence of given element in array

index : int = my_array.index(9)
print("Index of 7 :",index)

# Reverse the array
print(f"Array before reverse: {my_array}")

my_array.reverse()
print(f"Array after reverse : {my_array}")
```

5) Some other functions

```python
# Get the buffer.info and number of elements

my_array.buffer_info()

# Count the occurrence of given element in array
my_array.count(9)

# Converting to list
my_array.tolist()
```
